```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'retina'

import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mle_toolbox import load_result_logs, visualize
from neuroevobench.postprocess import filter_best, get_search_history
from neuroevobench.postprocess import plot_sensitivity, plot_history
```

```python
strategy_names = ['OpenES', 'PGPE', 'ARS', 'SNES', 'Sep_CMA_ES', 'LES',
                  'SimpleGA', 'SAMR_GA', 'GESMR_GA', 'LGA', 'BayesOptJAX']

plot_labels = ['OpenES', 'PGPE', 'ARS', 'SNES', 'SepCMA-ES', 'LES',
                  'SimpleGA', 'SAMR-GA', 'GESMR-GA', 'LGA', 'BayesOpt']

color_palette = sns.color_palette(
            "colorblind", n_colors=len(strategy_names)
        )
colors = dict(zip(strategy_names, color_palette))

metric = "test_eval_perf"
```

# Tuning Performance History Curves

```python
result_dirs = [
               "../experiments/fmnist_classify/",
               "../experiments/mnist_generate/",
               "../experiments/cifar/",
               "../experiments/ant/",
               "../experiments/fetch/",
               "../experiments/asterix/",
               "../experiments/breakout/",
               "../experiments/addition/",
               "../experiments/smnist/",
               ]

titles = [
    "FMNIST Classification [CNN]", "MNIST Generation [VAE]",
    "CIFAR10 Classification [CNN]",
    "Ant Control [MLP]", "Fetch Control [MLP]",
    "Asterix Control [CNN]", "Breakout Control [CNN]",
    "Addition Regression [GRU]", "SMNIST Classification [LSTM]",
]

perf_labels = ["Test Accuracy", "Neg. Test Loss", "Test Accuracy",
               "Episode Return", "Episode Return", "Episode Return", "Episode Return",
               "Neg. Test Loss", "Test Accuracy"]

fig, axs = plt.subplots(3, 3, figsize=(20, 17))

for i, dir in enumerate(result_dirs):
    eo_dir = os.path.join(dir, "search/")
    meta_log, hyper_log = load_result_logs(eo_dir)
    env_results = {}
    for strategy in strategy_names:
        if strategy not in ["BayesOptJAX", "LES"]:
            env_results[strategy] = get_search_history(hyper_log, meta_log, strategy, metric)
        if strategy == "LES":
            les_dir = os.path.join(dir, "search_les/")
            meta_l, hyper_l = load_result_logs(les_dir)
            env_results[strategy] = get_search_history(hyper_l, meta_l, strategy, metric)
        if strategy == "BayesOptJAX":
            if dir not in ["../experiments/mnist_generate/"]:
                bo_dir = os.path.join(dir, "search_bo/")
                meta_log, hyper_log = load_result_logs(bo_dir)
                env_results[strategy] = get_search_history(hyper_log, meta_log, strategy, metric)

    plot_history(env_results, fig, axs.flatten()[i],
                title=titles[i],
                ylabel=perf_labels[i],
                plot_ylabel=True, #(i==0 or i==3 or i==6),
                plot_xlabel=(i==len(axs.flatten())-3 or i==len(axs.flatten())-2 or i==len(axs.flatten())-1),
                colors=colors,
                plot_legend=i==0,
                plot_labels=plot_labels)

# fig.suptitle("Random Search: Best Performance Across Iterations", y=0.97, fontsize=45)
fig.tight_layout()
```
